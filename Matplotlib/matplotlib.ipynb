```python
from matplotlib import markers
import matplotlib.pyplot as plt
%matplotlib inline

value1 = [5,8,9,4,1,6,7,2,3,8]
value2 = [8,3,2,7,6,1,4,9,8,5]
plt.plot(range(1,11),value1,c='r',lw=1,ls='--',marker='<')
plt.plot(range(1,11),value2,c='b',lw=2,ls='-.',marker='o')
plt.xlabel('Roll No')
plt.ylabel('CPI')
plt.annotate(xy=[6.2,1],s='lowest CPI')
plt.legend(['cx','cy'],loc=4)
plt.show()
```
