```python
import numpy as np
a = np.array([1, 2])
b = np.array([2, 2])
np.less(a,b)

# OR >>> np.less([1,2],[2,2])
#    array([ True, False])

a = np.array([4, 2])
b = np.array([2, 2])
np.greater(a,b)

np.less_equal([4, 2, 1], [2, 2, 2])
np.greater_equal([4, 2, 1], [2, 2, 2])
```
